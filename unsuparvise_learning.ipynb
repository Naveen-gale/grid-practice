```python
import pandas as pd
```

```python
df = pd.read_csv("k-mean.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
X = df[
    [
        "CGPA",
        "DSA",
        "Aptitude",
        "Projects"
    ]
]
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
```

```python
x_scale= scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

wcss = []

for k in range(1,11):

    model = KMeans(
        n_clusters=k,
        random_state=42
    )

    model.fit(x_scale)

    wcss.append(model.inertia_)
```

```python
plt.plot(range(1,11), wcss)
plt.xlabel("K")
plt.ylabel("WCSS")
plt.show()
```

```python
model = KMeans(
    n_clusters=3,
    random_state=42
)

clusters = model.fit_predict(
    x_scale
)

df["Cluster"] = clusters
```
